```python
data = open('regex_text.txt')
rdata = data.read()
rdata
```

```python
import re

#URLs
regex = r"https?://(..............................)"
test_str = rdata
matches = re.finditer(regex, test_str,flags = re.MULTILINE)
for matchNum, match in enumerate(matches, start=1):
    
    print ("Match {matchNum} was found at {start}-{end}: {match}".format(matchNum = matchNum, start = match.start(), end = match.end(), match = match.group()))
    
```

```python
#finding the similarity
from itertools import combinations
from sklearn.metrics.pairwise import cosine_similarity
```

```python
lines = ['students of NLP are motivated but not interested','Students of NLP course are good in programming','students of nlp are motivated and good programmers']
vc = CountVectorizer()
Vlines1 = vc.fit_transform(lines)
pd.DataFrame(Vlines1.toarray(),columns=vc.get_feature_names())
```

```python
cv = CountVectorizer(stop_words = "english")
Vlines2 = cv.fit_transform(lines).toarray()
pd.DataFrame(Vlines2,columns=cv.get_feature_names())
```

```python
#finding the similarity
from itertools import combinations
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# list all of the combinations of 5 take 2 as well as the pairs of phrases
pairs = list(combinations(range(len(lines)),2))
combos = [(lines[a_index], lines[b_index]) for (a_index, b_index) in pairs]
combos
# calculate the cosine similarity for all pairs of phrases and sort by most
results = [cosine_similarity([Vlines2[a_index]], [Vlines2[b_index]]) for (a_index,
b_index) in pairs]
sorted(zip(results, combos), reverse=True)
```
